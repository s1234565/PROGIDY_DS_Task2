# **Exploratory Data Analysis(EDA)**

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.set(style="ticks")
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Embarked', data=df)
plt.title('Gender by Embarked')
plt.show()
```

```python
df.groupby(['Sex', 'Parch'])['Parch'].count()
```

```python
sns.countplot(x='Sex', hue='Parch', palette='PuRd', data=df)
plt.title('Parch Count by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Pclass', hue='Pclass', palette='magma', data=df)
plt.title('Pclass by Passenger Class')
plt.show()
```
